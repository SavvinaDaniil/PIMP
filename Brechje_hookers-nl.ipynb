```python
from urllib.request import urlopen

hookers_nl = "https://www.hookers.nl/forum/raamprostitutie/nederland/amsterdam"

f = urlopen(hookers_nl)
myfile = f.read()
```

```python
from bs4 import BeautifulSoup
from json import loads

soup = BeautifulSoup(myfile)
text = soup.get_text()

#find the values describing the reviews
json_alphabetic = soup.findAll('alphabetic')[0]

#extract the dict from the :topics attribute and delete string. 
dict_hookers = loads(json_alphabetic.attrs[':topics'][14:-1])
```

```python
import pandas as pd
df_hookers = pd.DataFrame(list(dict_hookers.values()))
df_hookers.head()
```

```python
df_hookers.street.unique()
```

```python

from collections import Counter

#count how many times the unique streets occur
counter_street = collections.Counter(df_hookers.street)

#make empty lists to store data
hookers = []
number_of_hookers = []
hooker_id = []

#look for each unique street
for i in counter_street.keys():
    
    #give the hookers' name if the street where the hooker is, is similar to the unique street
    hookers.append(df_hookers.name[df_hookers.street == i])
    hooker_id.append(df_hookers.nodeid[df_hookers.street == i])
    
#look into the names of the hookers we just found
for y in hookers:
    
    #add the number of hookers per location to a variable
    number_of_hookers.append(len(y))
    
df_new = pd.DataFrame()

#make a new column in the new dataframe with the number of hookers
df_new['number_of_hookers'] = number_of_hookers

#make the index the unique street names
df_new.index = counter_street.keys()

df_new['hookernames'] = hookers
df_new['hooker_id'] = hooker_id
    
```
